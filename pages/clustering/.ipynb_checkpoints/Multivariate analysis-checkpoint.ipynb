```r

# What is the dynamics range of the data?
2^(max(QuantTable)-min(QuantTable))



# Correlations between samples, replicates should show higher correlations
library(lattice)
levelplot(cor(QuantTable))

# Why is there now clear picture? -> Ratios and only few proteins changing

# create Ensembl set
eQuantTable <- QuantTable
rownames(eQuantTable) <- rQuantTable[,1]

# write into file for VSClust
write.csv(eQuantTable,"../../../Data/PMC6453966_VSClust_in.csv")

# PCA for some testing
pca.out <- princomp((QuantTable), cor = T)

#scoring plot
plot(pca.out$loadings, pch=16, col=rep(1:9,each=5) )


# heat map of the most changing proteins between  the  first 2


# Calculate means of each cancer subtype
AvQuant <- NULL
for (i in 1:5)
  AvQuant  <- cbind(AvQuant, rowMeans(QuantTable[,(i-1)*9+(1:9)]))
hist(AvQuant[,1] - AvQuant[,2], 100)  
# Most proteins tightly controlled

# What happens with the 100 most different proteins?
mostDiffInd <- names(sort(abs(AvQuant[,1] - AvQuant[,2]), decreasing = T)[1:100])
heatmap(AvQuant[mostDiffInd,], Colv=NA)

# Now look on the replicate level. Do you still have the same impression of significant changes?
heatmap(as.matrix(QuantTable[mostDiffInd,]), Colv = NA)

# Take e.g. MUC5B
plot(t(QuantTable["MUC5B",]), pch=16, col=rep(1:5,each=9))

# So we need statistics!
```
